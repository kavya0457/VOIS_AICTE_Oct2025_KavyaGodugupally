```python
# Airbnb Hotel Booking Analysis - Kavya

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Upload dataset
from google.colab import files
uploaded = files.upload()

import io
# Read the uploaded Excel file using pd.read_excel
file_name = list(uploaded.keys())[0] # Get the name of the uploaded file
df = pd.read_excel(io.BytesIO(uploaded['hotel_bookings.csv']))

# Preview data
df.head()
```

```python
import numpy as np

df['positive_review'] = np.where(df['review rate number'] >= 4, 1, 0)
id_rate = df.groupby('host_identity_verified')['positive_review'].mean().sort_values(ascending=False)
print(id_rate)
import matplotlib.pyplot as plt

id_rate.plot(kind='bar', color='skyblue')
plt.ylabel('Average Positive Review')
plt.title('Average Positive Review by Host Verification')
plt.show()

```
